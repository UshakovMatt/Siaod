```python
n = int(input())
x = list(map(int, input().split()))
m = max(x) + 1
ans = [0] * m
for a in x:
    ans[a] += 1
for i in range(2,m):
    ans[i] = max(ans[i-1], ans[i-2] + ans[i]*i)
print(max(ans))
```

```python
n,m = [int(i) for i in input().split(" ")]
tree={i:{} for i in range(1,n+1)}
a=[int(i) for i in input().split(" ")]
sz=[0 for i in range(0,n)]
for _ in range(n-1):
    c,d = [int(i) for i in input().split(" ")]
    tree[c][d]=1
    tree[d][c]=1
    sz[c-1]+=1
    sz[d-1]+=1
    visited= [False for i in range(0,n)]
q=[1]
count=0
while(q!=[]):
    k=q.pop(0)
    visited[k-1]=True
    for i in tree[k]:
        if not visited[i-1]:
        # q.append(i)
            if a[i-1]!=0:
                a[i-1]+=a[k-1]
            if a[i-1]<=m and sz[i-1]>1:
                q.append(i)
            if sz[i-1]==1 and a[i-1]<=m:
                # print(i,a[i-1])
                count+=1
print(count)
```

```python
def poweredADD(arr):
    m = arr[0]
    add = 0
    x = 0
    for i in arr:
        while m > i + add:
            x += 1
            add += 2**(x-1)
        m = max(m, i)
    print(x)

def main():
    t = int(input())
    for i in range(t):
        input()
        arr = list(map(int, input().split()))
        poweredADD(arr)

main()
```

```python

from sys import *
input = stdin.buffer.readline
n , m = map(int,input().split())
l = list(map(int, input().split()))
d = [0]*(n+1)
for i in range(m):
    x, y = map(int,input().split())
    d[x-1] += 1
    d[y] -= 1
for i in range(1,n):
    d[i] += d[i-1]
l.sort(reverse=True)
d.sort(reverse=True)
ka = 0
for i in range(n):
    ka += d[i]*l[i]
print(ka)
```
